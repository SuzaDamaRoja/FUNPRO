##ESTRUCTURA PARA/for
La estructura para se emplea cuando se sabe **cuántas veces** se repite una acción.

Sintaxis
```
for (variable) in range(VI,VF,pasos):
  La acción que se repite VF-VI
```

````python
#Hacer un programa que calcule la nómina de n empleado
nomina=0
n=int(input("Ingrese la cantidad de empleados: "))
for s in range(1,n+1,1):
  sueldo=int(input(f"Ingrese el salario del empleado {s}:"))
  nomina=nomina+sueldo #Acumulador
print(f"La nómina de {n} empleados es: {nomina}")
````

````python
#Hacer un programa que calcule la nómina de n empleado
nomina=0
n=int(input("Ingrese la cantidad de empleados: "))
for s in range(n):
  sueldo=int(input(f"Ingrese el salario del empleado {s+1}:"))
  nomina=nomina+sueldo #Acumulador
print(f"La nómina de {n} empleados es: {nomina}")
````

````python
#promedio de pares y suma de impares
#50 números cualesquiera
impar=0
pares=0
n=0

for i in range(53):#Quiero que se repita una cantidad de veces, en este caso, 10
  num=int(input("Ingrese un número: "))
  if num%2==0:
    pares=pares+num
    n=n+1
  else: impar=impar+num

print(f"La suma de los números impares es {impar}")
print(f"El promedio de los números pares es {pares/n}")
````
